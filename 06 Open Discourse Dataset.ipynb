```python
import spacy

nlp = spacy.load("de_core_news_sm")
```

```python
print(nlp)
```

## Preprocessing

```python
# But there are a couple of empty speeches in the dataset:

for i in range(len(opendiscourse.values)):
    speech = opendiscourse["speechContent"][i]
    if isinstance(speech, str) is not True:
        print(i, speech)
```

```python
# Let's look at one

opendiscourse.iloc[349487]
```

```python
# So now we have to get rid of all rows with faulty speeches:

opendiscourse = opendiscourse.dropna(subset = ['speechContent'])

# Note: we altered the amount of rows in our dataset. Their indexes however still resemble the old format, 
# which often causes problems. Therefore a reset is needed. 
# Therefore, let's reset our index:

opendiscourse.reset_index(drop=True)

```

```python
# test, if we successfully removed the NaN-entries:

len(opendiscourse)
```

```python
extra_stop = ["herr", "präsident", "frau", "damen", "herren"] # adapt if necessary
full_corpus = [] # for the final tokenized corpus

for i in range(100): # going through the first 100 entries
    
    # Please note: as there are almost 1 Million entries in the dataset, it runs quite long.
    # It makes sense to further filter the dataset, 
    # e.g. restricting it to specific years only (fitting your research question).
    # You can find the years interesting for your analysis by translating the id-information in the opendiscourse dataframe 
    # in e.g. 'session' or 'electoralTerm' with the information available in the other files available at opendiscourse.net
    # This allows us then to restrict the dataset on specific years
    
    speech = opendiscourse['speechContent'].iloc[i].replace('\n', ' ') # iterate through column 'speechContent' on i. row 
    # and replace all trailing new-line commands there with whitespace
    
    #print(i, speech)
    
    tokenized_speech = nlp(speech) # feed to our German language model
    
    speech_corpus = []
    for t in tokenized_speech:
        if not t.is_stop and not t.is_punct and not t.like_num and t.lemma_.lower() not in extra_stop:
            speech_corpus.append(t.lemma_.lower())

    full_corpus.append(speech_corpus)
    
```

```python
print(full_corpus)
```
