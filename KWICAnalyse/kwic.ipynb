```python
#schlagwörter aus dem topic modeling
# 2014: aufwachsen, jung, Kind, bleiben, Eltern, doppelt, Optionspflicht
# 2019: terroristisch, Terrormiliz, entziehen, Verlust, IS, Syrien
# 2021: Nationalsozialismus, Demut, Scham, verfolgter, Eltern, Großeltern, Nachkommen, Wiedergutmachung, Versöhnung, Unrecht
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# lade csv datei
data = pd.read_csv(r"C:\Users\Momo\s2014063.csv") 
#erstelle dataframe
df = pd.DataFrame(data)

# erstellen des tf-idf vectorizer
tfidf_vectorizer = TfidfVectorizer()

# umwandeln der datei
tfidf_matrix = tfidf_vectorizer.fit_transform(df['speechContent'])

# tf-idf matrix wird zu DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# anzeigen des tf-idf dataframe
print(tfidf_df)

# kwic analyse
keyword = "jung"  # manuelle eingabe des keywords
keyword_index = tfidf_df.columns.get_loc(keyword)

# herausfiltern der sätz, die das keyword enthalten
kwic_data = df['speechContent'].str.split(expand=True)
kwic_data['TF-IDF'] = tfidf_df[keyword]

# ergebnisse der kiwc analysis
print(kwic_data)
```
