```python
import pandas as pd
import json
import spacy
#import re
#import torch
import spacy_transformers
#import nltk
from spacytextblob.spacytextblob import SpacyTextBlob
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
nlp = spacy.load('de_core_news_sm')
nlp.add_pipe('spacytextblob')
```

```python
nlp.pipeline
```

```python
# json datei laden
with open(r"C:\Users\Momo\Desktop\Staatsangehoerigkeit - Kopie\Data19PlenarDocs\DrittesGesetz\sitzung16Mai.json") as data:
    plenar16 = json.load(data)
```

```python
#schleifen, um elemente aus list plenar16 zu filtern und in beitraege zu speichern
beitraege = []

for i in range(3, len(plenar16),5):
    beitraege.append(plenar16[i])

for beitrag in beitraege:
    print("_____________________________________________________________________")
    print(beitrag)
```

```python
# analysiert jeden satz auf sentiment
for beitrag in beitraege:
    doc = nlp(beitrag)
    for span in doc.sents:
        blob = TextBlob(span.text, analyzer = NaiveBayesAnalyzer())
        print(span.text, blob.sentiment)
```

```python
# polaritöt des gesamten text
doc._.polarity
```

```python
# subjektivität des gesamten text
doc._.subjectivity
```

```python
# zusammenfassung der ergebnisse für den gesamten text?
doc._.assessments
```
